- Importação da biblioteca pandas, criação do DataFrame e eliminação das duplicatas

```python
import pandas as pd
df = pd.read_csv('datasets/kc_house_data.csv')
df.drop_duplicates(subset='id', keep='last', inplace=True)
```

- Convertendo a coluna 'date' em datetime

```python
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
```

## Questão 1: Crie uma coluna house_age que indique se a casa é nova ou velha a partir da data de venda
- Casas vendidas a partir de 2014 serão classificadas como novas e antes disso como velhas

```python
df['house_age'] = None
df.loc[df.date >= '2014-01-01', 'house_age'] = 'new_house'
df.loc[df.date < '2014-01-01', 'house_age'] = 'old_house'
```

- Para conferir, será mostrada aa contagem de valores para a coluna house_age

```python
df['house_age'].value_counts()
```

- Para o dataframe em análise, todos os imóveis são classificados como new_house, uma vez que o arquivo kc_house_data.csv possui apenas casas com datas a partir de 2014-01-01

## Questão 2: Crie uma coluna dormitory_types que classifique os imóveis a partir do número de quartos
- Imóveis com 1 quarto: Studio;
- Imóveis com 2 quartos: Apartment;
- Imóveis com mais de 2 quartos: house.

```python
df['dormitory_types'] = None
df.loc[df.bedrooms == 1, 'dormitory_types'] = 'studio'
df.loc[df.bedrooms == 2, 'dormitory_types'] = 'apartment'
df.loc[df.bedrooms > 2, 'dormitory_types'] = 'house'
```

- Para conferir, será mostrada a contagem de valores para a coluna 'dormitory_types'

```python
df['dormitory_types'].value_counts()
```

## Questão 3: Crie a coluna condition_type  para classificar os imóveis de acordo com sua condição.
- Imóveis com condição menor ou igual a 2: Bad;
- Imóveis com condição entre 3 e  4: Regular;
- Imóveis com condição 5: Good.

```python
df['condition_type'] = None
df.loc[df.condition <= 2, 'condition_type'] = 'bad'
df.loc[(df.condition == 3) | (df.condition == 4), 'condition_type'] = 'regular'
df.loc[df.condition == 5, 'condition_type'] = 'good'
```

- Para conferir, será mostrada aa contagem de valores para a coluna contidion_type

```python
df['condition_type'].value_counts()
```

## Questão 6: Converta a coluna yr_built para datetime
- Para isso, será necessário utilizar o parâmetro format='%Y' para indicar que os valores da coluna equivalem ao ano.

```python
df['yr_built'] = pd.to_datetime(df['yr_built'], format="%Y")
```

## Questão 7: Converta a coluna yr_renovated para datetime
- Além de ser necessário indicar pelo format que os valores da coluna equivalem ao ano, é necessário utilizar o parâmetro errors='coerce' que faz com que os valores que não consigam fazer a conversão se torne NaT (Not a Time)

```python
df['yr_renovated'] = pd.to_datetime(df['yr_renovated'], format="%Y", errors="coerce")
```

## Questão 18: Selecione as colunas id, date, price, floors, zipcode de 3 formas diferentes
- Pelos índices das linhas e colunas;
- Pelos índices das linhas e nomes das colunas;
- Pelos índices das linhas e pelo tipo booleano para as colunas.

```python
column = ["id", "date", "price", "floors", "zipcode"]
columnsnum = []
for i, col in enumerate(df.columns.to_list()):
    if col in column:
        columnsnum.append(i)

print(df.iloc[0:20, columnsnum])
print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print(df.loc[0:20, column])
print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
columns = []
for i, col in enumerate(df.columns.to_list()):
    if col in column:
        columns.append(True)
    else:
        columns.append(False)
    
print(df.loc[0:20, columns])
```

## Questão 19: Salve um arquivo csv com as colunas de índices entre 10 e 17

## Questão 20: Salve um arquivo html com um mapa que mostre os imóveis com tamanhos distintos de acordo com o preço

```python
import plotly.express as px

df_map = df[['id', 'lat', 'long', 'price', 'condition', 'date']]

mapa = px.scatter_mapbox(df_map, lat='lat', lon='long', hover_name='id', hover_data=['price'],
                        color='condition', color_continuous_scale=px.colors.sequential.Viridis, zoom=9, size='price', size_max=15)
mapa.update_layout(mapbox_style='open-street-map', margin={'l': 0, 't': 0, 'r': 0, 'b': 0})

mapa.show()

mapa.write_html('datasets/map.html')
```
